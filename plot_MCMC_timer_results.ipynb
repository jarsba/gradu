```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
```

```python
CURRENT_FOLDER = os.path.dirname(os.path.abspath(__name__))
RESULTS_FOLDER = os.path.join(CURRENT_FOLDER, "results")
```

```python
df = pd.read_csv(f"{RESULTS_FOLDER}/napsu_MCMC_time_vs_epsilon_comparison.csv")
```

```python
df_with_task = df.query('task=="MCMC (LA)" or task=="MCMC"')
```

```python
g = sns.catplot(data=df_with_task, kind="bar", x="epsilon", y="timedelta", col="LA", errorbar=None)
g.set_axis_labels("Epsilon", "Time (s)")
g.set_titles("Laplace approximation: {col_name}")
plt.show()
```

```python
df_with_LA_task = df.query('task=="Laplace approximation"')
```

```python
g = sns.catplot(data=df_with_LA_task, kind="bar", x="epsilon", y="timedelta", errorbar=None)
g.set_axis_labels("Epsilon", "Time (s)")
plt.show()
```
